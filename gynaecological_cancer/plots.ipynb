```python
import pickle
```

```python
with open('metrics/metric_dict_efficientNet_B0_finetuned_model.pkl','rb') as f:
    data= pickle.load(f)
```

```python
with open('metrics/metric_dict_TinyVGG_base_model.pkl','rb') as f:
    data1 = pickle.load(f)
```

```python
with open('metrics/metric_dict_resnet_18_finetuned_model.pkl','rb') as f:
    data2 = pickle.load(f)
```

```python
with open('metrics/metric_dict_vit_B16_finetuned_model.pkl','rb') as f:
    data3 = pickle.load(f)
```

```python
import numpy as np
import matplotlib.pyplot as plt

array = np.arange(0,100,1)
plt.plot(array, data['train_loss'], color='r', label='EfficientNet B0 finetuned')
plt.plot(array, data1['train_loss'], color='g', label='TinyVGG base trained')
plt.plot(array, data2['train_loss'], color='c', label='ResNet 18 finetuned')
plt.plot(array, data3['train_loss'], color='b', label='ViT B16 finetuned')

# Adding a title
plt.title('Training Loss Comparison', fontsize=16, fontweight='bold')

# Adding x and y labels
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Loss', fontsize=12)

# Adding a grid for better readability
plt.grid(True, linestyle='--', alpha=0.6)

# Adding a legend to distinguish the plots
plt.legend(loc='upper right', fontsize=10)

# Adjusting the ticks for better appearance
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Setting plot limits (optional, depending on your data)
plt.xlim(0, 100)
plt.ylim(0, max(max(data['train_loss']), max(data1['train_loss']), max(data2['train_loss']), max(data3['train_loss'])))

# Adding a background color (optional)
plt.gca().set_facecolor('whitesmoke')

# Displaying the plot
plt.show()
```

```python
plt.plot(array, data['test_loss'], color='r', label='EfficientNet B0 finetuned')
plt.plot(array, data1['test_loss'], color='g', label='TinyVGG base trained')
plt.plot(array, data2['test_loss'], color='c', label='ResNet 18 finetuned')
plt.plot(array, data3['test_loss'], color='b', label='ViT B16 finetuned')

# Adding a title
plt.title('Test Loss Comparison', fontsize=16, fontweight='bold')

# Adding x and y labels
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Loss', fontsize=12)

# Adding a grid for better readability
plt.grid(True, linestyle='--', alpha=0.6)

# Adding a legend to distinguish the plots
plt.legend(loc='upper right', fontsize=10)

# Adjusting the ticks for better appearance
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Setting plot limits (optional, depending on your data)
plt.xlim(0, 100)

# Adding a background color (optional)
plt.gca().set_facecolor('whitesmoke')

# Displaying the plot
plt.show()
```

```python
plt.plot(array, data['train_acc'], color='r', label='EfficientNet B0 finetuned')
plt.plot(array, data1['train_acc'], color='g', label='TinyVGG base trained')
plt.plot(array, data2['train_acc'], color='c', label='ResNet 18 finetuned')
plt.plot(array, data3['train_acc'], color='b', label='ViT B16 finetuned')

# Adding a title
plt.title('Training Accuracy Comparison', fontsize=16, fontweight='bold')

# Adding x and y labels
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Accuracy', fontsize=12)

# Adding a grid for better readability
plt.grid(True, linestyle='--', alpha=0.6)

# Adding a legend to distinguish the plots
plt.legend(loc='lower right', fontsize=10)

# Adjusting the ticks for better appearance
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Setting plot limits (optional, depending on your data)
plt.xlim(0, 100)
plt.ylim(0.2, 1)  # Accuracy is typically between 0 and 1

# Adding a background color (optional)
plt.gca().set_facecolor('whitesmoke')

# Displaying the plot
plt.show()
```
